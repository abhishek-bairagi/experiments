```python
# Generate synthetic devices

from datetime import datetime, timedelta

# Sample data for device generation
os_choices = ["ios", "macos", "windows"]
models = {
    "ios": ["iPhone 13", "iPhone 14", "iPad Pro"],
    "macos": ["MacBook Air M1", "MacBook Pro M2", "iMac 2023"],
    "windows": ["Dell XPS 13", "HP EliteBook", "Surface Pro 9"]
}
os_versions = {
    "ios": ["16.4", "17.1", "17.3"],
    "macos": ["13.5", "14.0", "14.2"],
    "windows": ["10.0.19045", "11.0.22621", "11.0.22000"]
}

def generate_timestamp(days_ago):
    return (datetime.now() - timedelta(days=random.randint(1, days_ago))).isoformat()

# Generate devices
devices = []
device_id = 1
for user in range(1, 11):
    for _ in range(1):
        os_type = random.choice(os_choices)
        device = {
            "id": f"dev{device_id}",
            "os": os_type,
            "lastupdate": generate_timestamp(60),
            "os_version": random.choice(os_versions[os_type]),
            "ram": random.choice([4, 8, 16, 32]),  # GB
            "storage": random.choice([128, 256, 512, 1024]),  # GB
            "model": random.choice(models[os_type]),
            "issued_on": generate_timestamp(365)
        }
        devices.append(device)
        device_id += 1
```

```python
print(json.dumps(devices, indent=2))
```
