```python
import torch
from transformers import(
  AutoModelForQuestionAnswering,
  AutoTokenizer,
  pipeline
)
model_name = "sjrhuschlee/flan-t5-base-squad2"
```

```python

# a) Using pipelines
nlp = pipeline(
  'question-answering',
  model=model_name,
  tokenizer=model_name,
  # trust_remote_code=True, # Do not use if version transformers>=4.31.0
)

```

```python
policy ="""Sick Leave Policy
Our company prioritizes employee health and well-being by offering a comprehensive sick leave policy. Full-time employees are entitled to up to 10 days of paid sick leave annually, while part-time employees accrue sick leave proportionally based on their hours worked. Sick leave is designed to provide time off for personal illness, recovery, medical appointments, or to care for an immediate family member with a health condition.

Employees are expected to notify their supervisor or HR as soon as they become aware of the need for sick leave via mail. For absences lasting more than three consecutive days, a doctor’s note or relevant medical documentation may be required to ensure proper support and compliance. Unused sick leave cannot be carried over into the next calendar year, nor will it be paid out upon resignation or termination, except where legally mandated.

This policy ensures that employees can focus on their health without compromising their financial stability, fostering a supportive and productive workplace."""
```

```python
questions =  [
    "How many days of paid sick leave are full-time employees entitled to annually?",
    "Are part-time employees eligible for sick leave, and how is it calculated?",
    "What is the process for notifying the company about the need for sick leave?",
    "Is a doctor’s note required for all sick leave absences?",
    "What happens to unused sick leave at the end of the calendar year?"
]
```

```python
qa_input = {
'question': f'{nlp.tokenizer.cls_token}{questions[4]}',  # '<cls>Where do I live?'
'context': policy
}
res = nlp(qa_input)
print(res)
```
